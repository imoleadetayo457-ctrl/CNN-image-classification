```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("---Step 1: Loading the Pre-trained 'Brain'---")
model = VGG16(weights='imagenet')
print("VGG16 Model Loaded Successfully.")
```

```python
print("---Step 2: Preparing Our Custom Image---")
img_path = '/content/Screenshot 2026-04-29 194026.png'
img = image.load_img(img_path, target_size=(224, 224))

x = image.img_to_array(img)

x = np.expand_dims(x, axis=0)

x = preprocess_input(x)

print("Image Converted To Mathematical Array and Preprocessed.")
```

```python
print("\n---AI inference (Making Predictions)---")

print("Analyzing Viual Data.....")

prediction = model.predict(x)

results = decode_predictions(prediction, top=5)[0]
```

```python
print("\n---AI Classification Results---")
for result in results:
    label = results[1]
    confidence = result[2] * 100
    print(f"Prediction:{label} Confidence: {confidence:.2f}%)")
    plt.imshow(image.load_img(img_path))
    plt.axis('off')
    plt.show()
```
